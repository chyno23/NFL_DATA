```python
# Crear una lista para almacenar los datos reorganizados
partidos = []

# Procesar el DataFrame en bloques de 4 filas
for i in range(0, len(data_new), 2):  # Asumiendo que cada partido ocupa 2 filas
    if i + 1 < len(data_new):  # Asegurarse de que hay un par de filas
        # Extraer datos de cada bloque
        fecha = data_new.iloc[i, i*6]  # Fecha
        equipo_local = data_new.iloc[i, 1]  # Equipo local
        puntos_local = data_new.iloc[i, 2]  # Puntos local
        equipo_visitante = data_new.iloc[i + 1, 1]  # Equipo visitante
        puntos_visitante = data_new.iloc[i + 1, 2]  # Puntos visitante
        
        # Agregar los datos a la lista
        partidos.append([fecha, equipo_local, puntos_local, equipo_visitante, puntos_visitante])

# Crear un nuevo DataFrame a partir de la lista de partidos
df_partidos = pd.DataFrame(partidos, columns=['Fecha', 'Equipo Local', 'Puntos Local', 'Equipo Visitante', 'Puntos Visitante'])

# Mostrar el nuevo DataFrame
print("\nNuevo DataFrame reorganizado:")
df_partidos
```
